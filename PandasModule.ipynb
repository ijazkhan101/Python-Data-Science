```python
# 1st  iniings wickets
plt.figure(figsize = (10,10))
sns.countplot(x='FIRST INNINGS WICKETS',data=data,palette="winter")
plt.title("FIRST INNINGS WICKETS",fontsize=20)
plt.xlabel("Matches",fontsize=15)
plt.ylabel('WICKETS',fontsize=15)
plt.show()
```

```python
#Number of Matches win by a Team
plt.figure(figsize = (18,10))
sns.countplot(x='WINNER',data=data, palette='cool')
plt.title("MATCH WINNER  ",fontsize=20)
plt.xticks(rotation=50)
plt.xlabel("Teams",fontsize=15)
plt.ylabel("No of wins",fontsize=15)
plt.show()
```

```python
# match results
Win=data.WINNER.value_counts()
labels=np.array(Win.index)
sizes = Win.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# we will plot pie chart on Toss decision
Toss=data['TOSS DECISION'].value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss result')
plt.axis('equal')
plt.show()
```

```python
# we will print winner season wise
final_matches=data.drop_duplicates(subset=['WINNER'], keep='last')

final_matches[['VENUE','WINNER']].reset_index(drop=True).sort_values('WINNER')
```

```python
# we will plot pie chart on Winning percentage in final
match = final_matches['WON BY'].value_counts()
labels=np.array(Toss.index)
sizes = match.values
colors = ['gold', 'purple']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result')
plt.axis('equal')
plt.show()
```

```python
Toss=final_matches['TOSS DECISION'].value_counts()
labels=np.array(Toss.index)
sizes = Toss.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Toss Result')
plt.axis('equal')
plt.show()
```

```python
# we will print name of top player in T20 world cup
plt.figure(figsize = (18,10))
top_players = data['PLAYER OF THE MATCH'].value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,10])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="hsv");
plt.show()
```
